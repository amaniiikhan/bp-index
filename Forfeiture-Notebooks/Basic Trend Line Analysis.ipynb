```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("download.csv")
```

```python
df.groupby("")
```

```python
df = df.dropna(subset=['date', 'amount'])
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
described = df.groupby(df.date.dt.year)['amount'].describe()
```

```python
described['total'] = df.groupby(df.date.dt.year)['amount'].sum()
```

```python
sns.lineplot(x=described.index, y=described['total'], palette='pastel')
plt.title("Amount Collected From Forfeiture")
plt.xlabel("Year")
plt.ylabel("Amount (Millions $)")
```
