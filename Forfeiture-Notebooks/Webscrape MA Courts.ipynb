```python
soup=BeautifulSoup(response.content)
```

```python
table = soup.find("table", attrs={"class": "tableResults"})
table_head = table.thead.find_all("th")  # contains 2 rows


headings = []
for i in range(len(table_head)):
    headings.append(table_head[i].string)
headings
```

```python
table_data = table.tbody.find_all("tr")
data = []
for tr in table_data: # find all tr's from table's tbody
    t_row = {}
    # Each table row is stored in the form of
    # t_row = {'Rank': '', 'Country/Territory': '', 'GDP(US$million)': ''}

    # find all td's(3) in tr and zip it with t_header
    for td, th in zip(tr.find_all("td"), headings): 
        t_row[th] = td.text.replace('\n', '').strip()
    data.append(t_row)
```

```python
df = pd.DataFrame(columns=headings, data=data)
#df.loc[df['Initiating Action'] == "Forfeiture, G.L. c. 94C, ยง 47"]
```
